```python
#Membuat bentuk Geometri 2D dari OpenCV. Diawali dengan pembuatan black image
#dengan tipe data int16.

black_img = np.zeros(shape=(512,512,3), dtype=np.int16)
plt.imshow(black_img)
```

```python
#perhatikan koordinat titik2 pt1 dan pt2
cv.rectangle(black_img, pt1=(384,0), pt2=(510,128), color=(255,0,0), thickness=10)
plt.imshow(black_img)
```

```python
#Selanjutnya ditambah menambahkan bentuk persegi sesuai koordinat pt1 dan pt2
#yang tertulis pada kode program.

cv.rectangle(black_img, pt1=(200,200), pt2=(300,300), color=(0,255,0), thickness=10)
plt.imshow(black_img)
```

```python
#Tahap selanjutnya ditambah menambahkan bentuk lingkaran sesuai radius yang
#tertulis pada kode program.

cv.circle(black_img, center=(100,100), radius=50, color=(0,0,255), thickness=5)
plt.imshow(black_img)
```

```python
#Kemudian dilakukan penambahan garis sesuai koordinat pt1 dan pt2 sebagai berikut.
cv.line(black_img, pt1=(0,0), pt2=(512,512), color=(255,255,0), thickness=5)
plt.imshow(black_img)
```

```python
#Penambahan text dengan font yang telah tertulis dengan ukuran yang sudah
#ditentukan.
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img, text='PCVK', org=(10,500), fontFace=font, fontScale=4, color=(255,255,255), thickness=5, lineType=cv.LINE_AA)
plt.imshow(black_img)
```

```python
black_img2 = np.zeros(shape=(512,512,3), dtype=np.int32)
plt.imshow(black_img2)
```

```python
#Berikut adalah kode program untuk inisialisasi NumPy array dengan tipe data int32
vertices = np.array([[100, 300], [200,200], [400, 300], [200,400]], dtype=np.int32)
vertices
```

```python
pts = vertices.reshape((-1,1,2)) #nilai 2 untuk menunjukkan bahwa tiap titik dibuat 3 channel yang mewakili R, G, dan B pts
pts
```

```python
# Penambahan polyline pada black image kedua yang telah dibuat.
cv.polylines(black_img2, [pts], isClosed=True, color=(255,0,255), thickness=5)
plt.imshow(black_img2)
```

```python
#Dengan menggunakan figsize, perhatikan apakah ukuran image pixelnya juga berubah?
import matplotlib.pyplot as plt
import cv2

# Baca gambar
img = cv2.imread('/content/drive/MyDrive/PCVK/Images/kitten.jpg')  # Ganti dengan nama file Anda
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Menampilkan dengan figsize berbeda
plt.figure(figsize=(4,4))
plt.imshow(img_rgb)
plt.title("Figsize 4x4")
plt.show()

plt.figure(figsize=(8,8))
plt.imshow(img_rgb)
plt.title("Figsize 8x8")
plt.show()

print("Ukuran asli pixel:", img_rgb.shape)


```

Penjelasan : bisa diatur dengan parameter (x, y) dalam figsize

```python
#Tampilkan image dalam channel Red-Blue dan Green-Blue saja!
R, G, B = cv2.split(img_rgb)

# Red-Blue
RB = cv2.merge([R, B, B*0])
plt.imshow(RB)
plt.title("Red + Blue")
plt.show()

# Green-Blue
GB = cv2.merge([G, B, B])
plt.imshow(GB)
plt.title("Green + Blue")
plt.show()
```

```python
# Tampilkan image baris 20-115, kolom 25-120
crop_img = img_rgb[20:116, 25:121]  # slicing baris dan kolom
plt.imshow(crop_img)
plt.title("Baris 20-115, Kolom 25-120")
plt.show()
```

```python
#Tampilkan baris 5-30, semua kolom, channel Red saja
red_strip = R[5:31, :]  # hanya channel R
plt.imshow(red_strip, cmap='Reds')
plt.title("Baris 5-30, Channel Red")
plt.show()
```

```python
#Buat 5 kotak random
import numpy as np

# Membuat canvas kosong
canvas = np.zeros((300, 300, 3), dtype=np.uint8)

for _ in range(5):
    # Koordinat random
    x1, y1 = np.random.randint(0, 250, size=2)
    x2, y2 = x1 + np.random.randint(20, 50), y1 + np.random.randint(20, 50)
    color = tuple(np.random.randint(0, 255, size=3).tolist())
    cv2.rectangle(canvas, (x1, y1), (x2, y2), color, -1)

plt.imshow(canvas)
plt.title("5 Kotak Random")
plt.show()
```

```python
#Tampilkan image dengan posisi terbalik
flipped = cv2.flip(img_rgb, 0)  # 0 = vertikal, 1 = horizontal, -1 = keduanya
plt.imshow(flipped)
plt.title("Gambar Terbalik")
plt.show()
```

```python
#Rectangle dan Circle pada wajah
# Baca foto pribadi
face_img = cv2.imread('/content/drive/MyDrive/PCVK/Images/kitten.jpg')
face_rgb = cv2.cvtColor(face_img, cv2.COLOR_BGR2RGB)

# ===== Geser rectangle & circle =====
# Rectangle: titik awal (x1, y1), titik akhir (x2, y2)
cv2.rectangle(face_rgb, (80, 30), (230, 180), (255, 0, 0), 3)  # Geser kanan + atas

# Circle: center (cx, cy), radius
cv2.circle(face_rgb, (155, 80), 50, (0, 255, 0), 3)  # Geser kanan + atas

# Tampilkan hasil
plt.imshow(face_rgb)
plt.title("Rectangle & Circle Geser ke Atas dan Kanan")
plt.show()
```

```python
#Rectangle di sudut bawah kiri channel B
kitten = cv2.imread('/content/drive/MyDrive/PCVK/Images/kitten.jpg')
B_channel = kitten[:, :, 0]  # channel B

# Create a 3-channel image from the B channel for drawing
B_channel_display = cv2.cvtColor(B_channel, cv2.COLOR_GRAY2BGR)

# Gambar rectangle
cv2.rectangle(B_channel_display, (0, B_channel_display.shape[0]-50), (50, B_channel_display.shape[0]), (0, 0, 255), -1) # Changed color to blue for BGR

plt.imshow(B_channel_display)
plt.title("Rectangle di Sudut Bawah Kiri Channel B")
plt.show()
```
